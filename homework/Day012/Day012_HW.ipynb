### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)
soup=BeautifulSoup(r.text,"html5lib")
# print(soup.prettify)
for item in soup.find('',{'class':"part_list_2"}).find_all("h3"):
    print(item.em.text,":",item.a.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

# mylist =[{'date':d.find('span').text,'tag':d.find('em').text,'content':d.find('a').text.replace('\u3000',' ')} for d in soup.find(class_="part_list_2").find_all("h3")]

mydict=dict()
for item in soup.find(class_='part_list_2').find_all('h3'):
    tag=item.find('em').text
    if tag in mydict.keys():
        mydict[tag].append(item.find('a').text)
    else:
        mydict[tag]=list()
        mydict[tag].append(item.find('a').text)
for key, value in mydict.items() :
    print ('\n',key)
    for x in value:
        print(x)
```

### ③ 哪一個類別的文章最多

```python
count=0
name=""
for key,value in mydict.items():
    if(count<len(value)):
        count=len(value)
        name=key
print(name,count)
```
