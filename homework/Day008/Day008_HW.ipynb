### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'
r=requests.get(url)
r.encoding = 'utf-8'
response=r.text
print(response)
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')
print(type(response))
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')
from bs4 import BeautifulSoup
soup=BeautifulSoup(response,'html.parser')
print(type(soup))
print(soup.prettify())
```
